```python
def ma_mh_ma_lop_correct(row):
    row.ma_mh = row.ma_mh.upper()
    if pd.isnull(row.ma_lop):
        row.ma_lop = row.ma_mh
        return row
    
    row.ma_lop = row.ma_lop.upper()
    if not row.ma_lop.startswith(row.ma_mh):
        row.ma_lop = row.ma_mh
    return row
new_df = new_df.apply(ma_mh_ma_lop_correct, axis=1)
```

# fill null diem_qt, diem_th, diem_gk, diem_ck, diem_hp

```python
new_df.info()
```

```python
diem_columns = ['diem_qt', 'diem_th', 'diem_gk', 'diem_ck', 'diem_hp']
new_df.loc[trang_thai_mien, new_df.columns.isin(diem_columns)] = float('nan')
new_df.loc[trang_thai_mien, new_df.columns.isin(diem_columns)]
```

```python
new_df.loc[(~trang_thai_mien) & (new_df.isnull().any(axis=1)), diem_columns]
```

```python
new_df.loc[(~trang_thai_mien), diem_columns]
```

```python
def fill_row(row):
    fill_value = row.mean()
    return row.fillna(fill_value)

tmp = new_df.loc[(~trang_thai_mien) & (new_df.isnull().any(axis=1)), diem_columns]
new_df.loc[(~trang_thai_mien) & (new_df.isnull().any(axis=1)), diem_columns] = tmp.apply(fill_row, axis=1)
new_df.loc[~trang_thai_mien, diem_columns]
```
