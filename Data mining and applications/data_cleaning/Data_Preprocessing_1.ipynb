2.Phân tích vấn đề:
+ Xác định câu hỏi nghiên cứu hoặc kinh doanh: Bạn đang cố gắng giải 
quyết vấn đề gì trong thực tế? Bạn cần đo lường điều gì và sẽ đo lường 
nó như thế nào?
+ Thu thập dữ liệu thô: Thu thập các tập dữ liệu thô cần dùng để trả lời câu 
hỏi trên. Việc thu thập dữ liệu có thể đến từ các nguồn nội bộ (các hệ
thống quản lý, …) hoặc từ các nguồn khác bên ngoài (mạng xã hội, công 
cụ tìm kiếm, …).
+ Làm sạch dữ liệu: Làm sạch dữ liệu để chuẩn bị cho phân tích. Điều này 
thường liên quan đến việc:
 + Xóa dữ liệu trùng lặp và bất thường.
 + Điều chỉnh sự không nhất quán.
 + Chuẩn hóa cấu trúc và định dạng dữ liệu.
 + Xử lý khoảng trắng và các lỗi cú pháp khác.
+ Phân tích dữ liệu: Bằng cách thao tác dữ liệu với các kỹ thuật và công cụ
phân tích dữ liệu khác nhau. Tìm ra các xu hướng, mối tương quan, giá 
trị ngoại lệ và các biến thể để kể một câu chuyện Trong giai đoạn này, 
bạn có thể sử dụng khai thác dữ liệu để khám phá các mẫu trong cơ sở dữ
liệu hoặc phần mềm trực quan hóa dữ liệu để giúp chuyển đổi dữ liệu 
thành định dạng đồ họa dễ hiểu.
+ Giải thích kết quả phân tích: Từ đó, xác định dữ liệu đã trả lời câu hỏi ban 
đầu của bạn tốt như thế nào. Bạn có thể đưa ra khuyến nghị gì dựa trên
